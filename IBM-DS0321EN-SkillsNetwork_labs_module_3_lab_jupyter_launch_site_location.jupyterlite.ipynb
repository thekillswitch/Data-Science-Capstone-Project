```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
import folium
from folium.plugins import MarkerCluster

# Sample data with 'class' column (replace this with your dataset)
launch_sites = [
    {"site": "Site A", "lat": 28.562302, "lon": -80.577356, "class": 1},
    {"site": "Site B", "lat": 28.573255, "lon": -80.646895, "class": 0},
    # Add other launch sites with their coordinates and 'class' values here
    # {"site": "Site Name", "lat": latitude, "lon": longitude, "class": class_value},
    # ...
]

# Create a base map centered around a specific location
m = folium.Map(location=[28.5, -80], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(m)

# Define a function to determine marker color based on the 'class' value
def get_marker_color(class_value):
    return 'green' if class_value == 1 else 'red'

# Iterate through the launch sites and add markers to the MarkerCluster
for site in launch_sites:
    popup_text = f"{site['site']} - Class: {site['class']}"
    marker_color = get_marker_color(site['class'])
    folium.Marker([site['lat'], site['lon']], popup=popup_text, icon=folium.Icon(color=marker_color)).add_to(marker_cluster)

# Save the map as an HTML file
m.save("launch_sites_map_class_colored.html")
m
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    # Earth's radius in kilometers
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    # Calculate differences in coordinates
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    # Haversine formula to calculate distance
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    
    return distance

# Coordinates of the launch site and the coastline (replace with your actual coordinates)
launch_site_lat = 28.56367
launch_site_lon = -80.57163
coastline_lat = 28.5
coastline_lon = -80.6

# Calculate distance between launch site and coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create a folium.Map object
m = folium.Map(location=[28.5, -80.5], zoom_start=8)

# Add a marker for the selected closest coastline point
coastline_marker = folium.Marker([coastline_lat, coastline_lon], 
                                 icon=folium.DivIcon(icon_size=(20,20), icon_anchor=(0,0),
                                                     html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline)))

# Add the coastline marker to the map
m.add_child(coastline_marker)

# Display the map
m

```

```python
import folium
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    # Earth's radius in kilometers
    R = 6371.0
    
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    # Calculate differences in coordinates
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    # Haversine formula to calculate distance
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    
    return distance

# Coordinates of the launch site and the selected point (replace with your actual coordinates)
launch_site_lat = 28.56367
launch_site_lon = -80.57163
poi_lat = 28.6139  # Example closest city latitude
poi_lon = -81.2097  # Example closest city longitude

# Calculate distance between launch site and the point of interest (POI)
distance_poi = calculate_distance(launch_site_lat, launch_site_lon, poi_lat, poi_lon)

# Create a Folium map object
m = folium.Map(location=[28.5, -80.5], zoom_start=8)

# Create a marker for the point of interest showing the distance
poi_marker = folium.Marker([poi_lat, poi_lon], 
                           icon=folium.DivIcon(icon_size=(20,20), icon_anchor=(0,0),
                                               html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_poi)))

# Add the point of interest marker to the map
m.add_child(poi_marker)

# Create a PolyLine between launch site and the point of interest
polyline = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [poi_lat, poi_lon]], weight=1)

# Add the PolyLine to the map
m.add_child(polyline)

# Display the map
m
```
